```python
# Program 23: Prime Number Check Using For Loop

num = 7
count = 0

for i in range(1, num + 1):
    if num % i == 0:
        count += 1

if count == 2:
    print("Prime Number")
else:
    print("Not Prime Number")
```

```python
# Program 24: Palindrome Number Using While Loop

num = 121
temp = num
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# Program 25: Armstrong Number Using While Loop

num = 153
temp = num
total = 0

while num > 0:
    digit = num % 10
    total += digit ** 3
    num = num // 10

if temp == total:
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
```
